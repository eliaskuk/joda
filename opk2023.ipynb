#### Aineiston kerääminen ja esikäsittely - Datan muokkaaminen

Monesti data vaatii muokkaamista ennen kuin sitä voidaan alka hyödyntämään. Datassa saattaa olla tyhjiä arvoja, samat asiat saattavat olla kirjoitettu eri tavoin tai teksteissä voi olla kirjoitus virheitä. On siis erittäin tärkää, että ennen kuin dataa aletaan hyödyntämään, tulee se siivota. Seuraavaksi siivotaan testidata df test_df_01.

```python
# Tutkitaan, löytyykö test_data_01 tyhiä rivejä
print("age nulls:", test_df_01['age'].isnull().sum())
print("age_0 nulls:",test_df_01['age_o'].isnull().sum())
print("attractive nulls:",test_df_01['attractive'].isnull().sum())
print("attractive_o nulls:",test_df_01['attractive_o'].isnull().sum())
print("race nulls:",test_df_01['race'].isnull().sum())
print("race_o nulls:",test_df_01['race_o'].isnull().sum())
print("expected_num_matches nulls:",test_df_01['expected_num_matches'].isnull().sum())
print("match nulls:",test_df_01['match'].isnull().sum())
```

```python
# Poistetaan rivit, joilla age tai age_o on null
test_df_01 = test_df_01.dropna(subset=['age'])
test_df_01 = test_df_01.dropna(subset=['age_o'])
test_df_01 = test_df_01.dropna(subset=['attractive'])
test_df_01 = test_df_01.dropna(subset=['attractive_o'])
test_df_01 = test_df_01.dropna(subset=['expected_num_matches'])

# Lisätään race_o riveille "Other", jotka ovat tyhjiä
test_df_01["race"] = test_df_01["race"].fillna("Other")
test_df_01["race_o"] = test_df_01["race_o"].fillna("Other")

# Tehdään tarkistus
print("age nulls:", test_df_01['age'].isnull().sum())
print("age_0 nulls:",test_df_01['age_o'].isnull().sum())
print("attractive nulls:",test_df_01['attractive'].isnull().sum())
print("attractive_o nulls:",test_df_01['attractive_o'].isnull().sum())
print("race nulls:",test_df_01['race'].isnull().sum())
print("race_o nulls:",test_df_01['race_o'].isnull().sum())
print("expected_num_matches nulls:",test_df_01['expected_num_matches'].isnull().sum())
print("match nulls:",test_df_01['match'].isnull().sum())
```

```python
# Tarkastetaan, ettei datassa esiinny kirjoitusvirheitä tai duplikaatteja
unique_races = test_df_01['race'].unique()
print(unique_races)

unique_races_o = test_df_01['race_o'].unique()
print(unique_races_o)
```

Datassa ei esiinny kirjoitusvirheitä ja tyhjät arvot on poistettu. Nyt voidaan one hotata race ja race_o sarakkeet, jotta niitä voidaan hyödyntää paremmin koneoppimismalleissa.

```python
# Lisätään race_o arvojen perään '_o', jotta niille voidaan luoda omat sarakkeet
# Tämän koodin voi ajaa vain kerran tai syntyy error, silloin pitää test_df_01 ajaa uudelleen
test_df_01['race_o'] = test_df_01['race_o'].cat.add_categories(['Asian/Pacific Islander/Asian-American_o'])
test_df_01['race_o'] = test_df_01['race_o'].replace({'Asian/Pacific Islander/Asian-American': 'Asian/Pacific Islander/Asian-American_o'}, regex=True)

test_df_01['race_o'] = test_df_01['race_o'].cat.add_categories(['European/Caucasian-American_o'])
test_df_01['race_o'] = test_df_01['race_o'].replace({'European/Caucasian-American': 'European/Caucasian-American_o'}, regex=True)

test_df_01['race_o'] = test_df_01['race_o'].cat.add_categories(['Other_o'])
test_df_01['race_o'] = test_df_01['race_o'].replace({'Other': 'Other_o'}, regex=True)

test_df_01['race_o'] = test_df_01['race_o'].cat.add_categories(['Latino/Hispanic American_o'])
test_df_01['race_o'] = test_df_01['race_o'].replace({'Latino/Hispanic American': 'Latino/Hispanic American_o'}, regex=True)

test_df_01['race_o'] = test_df_01['race_o'].cat.add_categories(['Black/African American_o'])
test_df_01['race_o'] = test_df_01['race_o'].replace({'Black/African American': 'Black/African American_o'}, regex=True)
```

```python
test_df_01
```

```python
one_hot = pd.get_dummies(test_df_01['race'])
one_hot_o = pd.get_dummies(test_df_01['race_o'])

# yhdistetään one hotit test_df_01 kanssa
test_df_01 = pd.concat([test_df_01, one_hot], axis=1)
test_df_01 = pd.concat([test_df_01, one_hot_o], axis=1)

# Poistetaan sarakkeet race ja rece_o
test_df_01 = test_df_01.drop(['race', 'race_o'], axis=1)


test_df_01
```

```python
# Lasketaan jokaisen  yhdistelmän lukumäärä age ja attractive yhdistelmistä.
grouped = test_df_01.groupby(['age', 'attractive']).size().reset_index(name='count')

# Luodaan taulukko, jossa kukin yhdistelmä saa koon yhdistelmän esiintymien määrään mukaan.
sizes = [count for count in grouped['count']]

# Luodaan pistekaavio. Pistekoko määritetään yhdistelmien lukumäärän perusteella.
plt.scatter(grouped['age'], grouped['attractive'], s=sizes)
plt.title('Ulkonäkö iän funktiona')
plt.xlabel('Ikä')
plt.ylabel('Ulkonäkö')
plt.show()
```

## Datatiedeprojektin pohja - Koneoppiminen

Koneoppiminen on tietojenkäsittelytieteen ala, joka keskittyy kehittämään algoritmeja ja menetelmiä, joilla mahdollistetaan tietokoneiden oppiminen esimerkkidatan perusteella. Koneoppimisen ohjelmat pyrkivät löytämään malleja ja sääntöjä datan analysoinnin ja prosessoinnin avulla, jonka auttaa voidaan suorittaa monimutkaisia tehtäviä tehokkaammin ja tarkemmin kuin perinteiset ohjelmointitekniikat. Koneoppimisen kategorioihin kuuluvat ohjattu oppiminen, ohjaamaton oppiminen, puoliohjattu oppiminen ja vahvistettu oppiminen.  

Ohjatussa oppimisessa ohjelma oppii sille syötetystä esimerkkidatasta, jossa jokaiselle syötteelle on merkitty oikea vastaus. Valvomattomassa oppimisessa ohjelma pyrkii löytämään rakenteita ja kuvioita syötteestä ilman ennakkotietoa oikeista vastauksista. Puoliohjatussa oppiminen on välimuoto ohjatun ja ohjaamattoman oppimisen välillä, jossa hyödynnetään molempien oppimisen periaatteita. Ennustava malli on malli, joka on rakennettu koneoppimisalgoritmin ja piirteiden tai attribuuttien avulla koulutusaineistosta, jonka avulla voidaan ennustaa tuleva arvo hyödyntämällä koulutuaineistoa.  Ohjatut oppimisalgoritmit pyrkivät mallintamaan suhteita ja riippuvuuksia kohde-ennusteen tuotoksen ja syötteiden piirteiden välillä. Käytetyimmät ohjatun oppimisen algoritmityypit ovat luokittelu ja regressio. 

Luokittelualgoritmi on ohjatun oppimisen menetelmä, jonka tavoitteena on luokitella uusia "arvoja" tiettyihin luokkiin. Luokittelualgoritmi saa opetusdatana joukon esimerkkejä, joista jokainen sisältää piirteitä ja luokan, johon se kuuluu. Algoritmi käyttää näitä esimerkkejä oppiakseen tunnistamaan piirteitä, jotka ovat ominaisia kullekin luokalle. Luokittelualgoritmi voi olla esimerkiksi:
- kasvojentunnistusalgoritmi
- sairauden diagnosointialoritmi
- sähköpostien luokittelualgoritmi

Regressioalgoritmissa tavoitteena on ennustaa numeerinen arvo muuttujan perusteella. Regressioalgoritmit hyödyntää opetusdataa oppiakseen ennustamaan piirteitä, jotka ovat yhteydessä riippuvan muuttujan arvoon. Niitä voidaan hyödyntää monissa eri asioissa kuten: 
- Ennustava analytiikka
- Markkinointitutkimus
- Sääennuste
- Taloudellinen ennustaminen

Seuraavaksi käydään läpi esimerkki, jossa luokitellaan twiittejä niiden retweettausten määrien perusteella. Datana hyödynnetään Donald Trumpin tweettejä. Tarkoituksena on luokitella, saako Tweetti sen tekstin perusteella enemmän retweettausia kuin keskimäärin Donald Trumpin tweetit ovat saaneet.

```python
# Ladataan nltk kirjaston tokenit
nltk.download('punkt')
nltk.download('stopwords')
nltk.data.path.append("<path/to/nltk_data>")
```

```python
# Ladataan valittu data API:sta, dataset # 43449
# Ladattava data on muodossa ARFF, tehdään sitä pandas dataframe
dataset = openml.datasets.get_dataset(43449)
arff_data = dataset.get_data()
df = pd.DataFrame(arff_data[0])

df_tweet = pd.DataFrame(df[['text','retweets']])

# kaikki pienillä kirjaimilla
df_tweet['text'] = df_tweet['text'].str.lower()

# poistetaan erikoisemerkit
df_tweet['text'] = df_tweet['text'].apply(lambda x: re.sub(r'[^\w\s]', '', x))

# Poistetaan "Stop-wordit"
stop_words = set(stopwords.words('english'))
df_tweet['text'] = df_tweet['text'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stop_words]))

# Tokenisoidaan tweetit
tokens = nltk.word_tokenize(" ".join(df_tweet['text'].tolist()))
freq = nltk.FreqDist(tokens)

# Valitaan 20 käytetyintä sanaa ja muodostetaan niistä kuvaaja
top_words = freq.most_common(20)
words = [word[0] for word in top_words]
counts = [count[1] for count in top_words]

plt.bar(words, counts)
plt.xticks(rotation=90)
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.title('Top 20 Most Used Words')
plt.show()
```

```python
# Lasketaan keskiarvo uudelleen twiittauksista mallia varten
average_retweets = df_tweet['retweets'].mean()
print(average_retweets)
```

```python
# Luodaan sarake, onko enemmän vai vähemman uudelleen twiittausksia kuin keskiarvo
df_tweet['more_than_average'] = (df_tweet['retweets'] > average_retweets).astype(int)

# Jaetaan testaus ja koulutusdataan
X_train, X_test, y_train, y_test = train_test_split(df_tweet['text'], df_tweet['more_than_average'], random_state=42)

# Vektorisoidaan teksti
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Koulutetaan naive Bayes classifier
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_vec, y_train)

# Testataan mallin toimivuutta
y_pred = nb_classifier.predict(X_test_vec)

# Tulostetaan classification raportti
print(classification_report(y_test, y_pred))
```

Tuloksista voidaan lukea, että yleinen tarkkuus (accuracy) on 68%. Tarkkuusarvo (precision) luokalle 0 (twiitit, joilla on vähemmän uudelleen twiittauksia kuin keskiarvo) on 0.73, eli luokittelija ennusti 73% oikein niistä. Luokalle 1 (enemmän kuin keskiarvo) tarkkuusarvo 0.60, eli 60% oikein, jotka yli keskiarvon. Tunnistuskyky (recall) luokalle 0 on 0.75, ja luokalle 1 0.58, mikä tarkoittaa, että kaikista niistä twiiteistä, jotka todella olivat vähemmän uudelleen twiitattuja kuin keskiarvo, luokittelija ennusti oikein 75% ja 58% oikein todellisista yli keskiarvon olleista tweeteistä. F1 mittari tarjoaa yhdistetyn mittarin edellä mainituista mittareista. Support kertoo luokkien esiintymismäärät, eli paljonko yli ja alle keskiarvojen.
