Eli kuinka suuri osa tietyn etnisen taustan omaavista henkilöistä ovat käyneet eri etnisten taustojen kanssa pikadeitellä. Violinplot näyttää, miten datan vastanneet sukupuolet jakautuvat. Pylväsdiagrammissa hyödynneteään muokattua dataa test_df_01 ja violinplotissa dataframea df, sillä se pitää sisällään sukupuolet.

```python
import matplotlib.pyplot as plt

# Luodaan ryhmä, jossa race and race_o välillä ja jokaisen ryhmän koko. Nollataan indeksi.
grouped = test_df_01.groupby(['race', 'race_o']).size().reset_index(name='count')

# Muutetaan race indeksiksi ja race_o sarakkeeksi, arvot sarakkeelle otetaan grouped count sarakkeesta
pivoted = grouped.pivot(index='race', columns='race_o', values='count')

# Luodaan stacked pylväsdiagrammi
ax = pivoted.plot(kind='bar', stacked=True)
ax.set_xlabel('Race')
ax.set_ylabel('Count')
ax.legend(title='Dates with')
plt.show()
```

#### Koneoppiminen - Ohjattu oppiminen

Kuten yllä mainittiin, ohjattu oppiminen perustuu ohjelmalle opetettuun koulutusaineistoon. Ohjattua oppimista käytetään yleisesti ennustavan mallinnuksen tekemisessä. Ennustava malli on malli, joka on rakennettu koneoppimisalgoritmin ja piirteiden tai attribuuttien avulla koulutusaineistosta, jonka avulla voidaan ennustaa tuleva arvo hyödyntämällä koulutuaineistoa.  Ohjatut oppimisalgoritmit pyrkivät mallintamaan suhteita ja riippuvuuksia kohde-ennusteen tuotoksen ja syötteiden piirteiden välillä. Käytetyimmät ohjatun oppimisen algoritmityypit ovat luokittelu ja regressio. 

Luokittelualgoritmi on ohjatun oppimisen menetelmä, jonka tavoitteena on luokitella uusia "arvoja" tiettyihin luokkiin. Luokittelualgoritmi saa opetusdatana joukon esimerkkejä, joista jokainen sisältää piirteitä ja luokan, johon se kuuluu. Algoritmi käyttää näitä esimerkkejä oppiakseen tunnistamaan piirteitä, jotka ovat ominaisia kullekin luokalle. Luokittelualgoritmi voi olla esimerkiksi:
- kasvojentunnistusalgoritmi
- sairauden diagnosointialoritmi
- sähköpostien luokittelu algoritmi

Regressioalgoritmissa tavoitteena on ennustaa numeerinen arvo muuttujan perusteella. Regressioalgoritmit hyödyntää opetusdataa oppiakseen ennustamaan piirteitä, jotka ovat yhteydessä riippuvan muuttujan arvoon. Niitä voidaan hyödyntää monissa eri asioissa kuten: 
- Ennustava analytiikka
- Markkinointitutkimus
- Sääennuste
- Taloudellinen ennustaminen

Seuraavaksi käydään ohjattua oppimista koodiesimerkkien avulla läpi. Ensimmäiseksi luokittelu (logistinen regressio) ja sitten regressio (lineaarinen regressio).

```python
# Luodaan esimerkkit, miten luokittelua ja regressiota voidaan hyödyntää datatiedeprojektissa
# Aluksi muokataan data käytettävään muotoon

test_df_02 = df.loc[:,['age', 'age_o', 'attractive', 'attractive_o','match']]

print("age nulls:", test_df_02['age'].isnull().sum())
print("age_0 nulls:",test_df_02['age_o'].isnull().sum())
print("attractive nulls:",test_df_02['attractive'].isnull().sum())
print("attractive_o nulls:",test_df_02['attractive_o'].isnull().sum())
print("match nulls:",test_df_02['match'].isnull().sum())
```

```python
# Poistetaan rivit, joilla on null
test_df_02 = test_df_02.dropna(subset=['age'])
test_df_02 = test_df_02.dropna(subset=['age_o'])
test_df_02 = test_df_02.dropna(subset=['attractive'])
test_df_02 = test_df_02.dropna(subset=['attractive_o'])
```

```python
# Luokitellaan logistisen regression avulla data ja luodaan malli, joka ennustaa tuleeko match,
# kun tiedetään arvio omasta ulkonäöstä (1-10), arvio toisen ulkonäöstä (1-10) sekä molempien iät

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score

# Jaetaan data koulutus- ja testausdataksi. 30% testidataa
train, test = train_test_split(test_df_02, test_size=0.3)

# Erotellaan X ominaisuudet (ikä ja ulkonäkö) ja y tulokset (match: kyllä/ei) koulutus- ja testausdatasta
# Ekat sarakkeet on iät ja ulkonäkö, viimeinen on match
X_train = train.iloc[:, :-1]
y_train = train.iloc[:, -1]
X_test = test.iloc[:, :-1]
y_test = test.iloc[:, -1]

# Sovitetaan logistinen regressiomalli
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Ennustetaan matchit testidatasta
y_pred_01 = log_reg.predict(X_test)

# Arvioidaan mallin suorituskykyä
accuracy = accuracy_score(y_test, y_pred_01)
roc_auc = roc_auc_score(y_test, y_pred_01)

print("Accuracy:", round(accuracy,2))
print("ROC AUC:", round(roc_auc,2))
```

Nyt luotu malli arvioi noin 84% prosentin todennäköisyydellä oikein, tulleko match. ROC AUC arvo on n. 0,5 mikä on lähellä satunnaisarvon (0.5) ennustuskykyä. Tämä tarkoittaa, että mallin ennustuskyky on hyvin heikko ja se ei pysty erottamaan positiivisia ja negatiivisia havaintoja kovin hyvin.

```python
# Arvioidaan hyvännäköisyys iän perusteella käyttäen lineaarista regressiota.

from sklearn.linear_model import LinearRegression

# Jaetaan data opetus- ja testausjoukkoihin
X_train, X_test, y_train, y_test = train_test_split(test_df_02['age'], test_df_02['attractive'], test_size=0.2)

# Jaetaan data opetus- ja testausjoukkoihin
regressor = LinearRegression()

# Muokataan X_train 2D-taulukoksi
X_train = X_train.values.reshape(-1, 1)
regressor.fit(X_train, y_train)

# Ennustetaan testidataan kuuluvien henkilöiden "attractice" käyttäen opetettua mallia
X_test = X_test.values.reshape(-1, 1) # reshape X_test to 2D array
y_pred = regressor.predict(X_test)

# Arvioidaan mallin suorituskykyä R-neliöarvon avulla
score = regressor.score(X_test, y_test)
print("R-squared score: {:.2f}".format(score))
```

R-arvoksi tulee 0,01 mikä tarkoittaa sitä, että data ei sovi kovin hyvin lineaariseen regressioon. Jotta voidaan ymmärtään paremmin kyseisen mallin huono toimivuus, voidaan piirtää koordinaatistoon data ja arvioida tuloksia lisää sen perusteella.

Voidaan vielä tarkastaa muuttujien yhteys toisiinsa korrelaation avulla.

Korrelaatio on nyt lähellä nollaa, mikä tarkoittaa siis sitä, että iällä ja hyvännäköisyydellä ei ole korrelaatioita, eli näillä muuttujilla ei ole tilastollista yhtettä toistensa kanssa.
