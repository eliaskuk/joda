## Datatiedeprojektin pohja - Prosessi

Vaikka datatiede onkin vielä kokonaisuutena uudehko tieteenala, on sille muodostunut jo monia eri malleja ja toimintatapoja. Nämä helpottavat projetkien suunnittelua sekä mahdollistavat loogisen kokonaisuuden muodostamisen.  Esimerkiksi seuraavat prosessimallit ovat yleisesti hyödynnettyjä datatieteessä.
- CRISP-DM. Lyhynne CRISP-DM on sanoista CRoss Industry Standard Process for Data Mining. Prosessi pitää sisällään liiketoiminnan ymmärtämisen, datan ymmärtämisen, datan valmistelun, mallintamisen, arvioinnin, käyttöönoton ja tietenkin sen tärkeimmän - datan.
- SEMMA. Lyhenne tulee sanoista Sample, Explore, Modify, Model, Assess. Prosessimalli koostuu viidestä vaiheesta, jotka ovat näytteenotto, tutkiminen, muokkaaminen, mallintaminen ja arviointi.
- TDSP. Malli Team Data Science Process on Microsoftin kehittämä prosessimalli, ja se sisältää viisi vaihetta: liiketoiminnan ymmärtäminen, tietojen valmistelu, mallien rakentaminen, käyttöönotto ja arviointi.

Kyseisessä työssä hyödynnetään TDSP mallia, sillä se tarjoaa riittävän laajan kokonaisuuden prosessin eri osa-alueista. Malli pitää sisällään myös osa-alueita, joihin työssä ei juuri paljoa syvennytä, esimerkiksi data engineeringin puolella. 



#### Vaihe 1. 
Prosessimalli ja kyseinen työ alkaa vaiheesta liiketoiminnan ymmärtäminen. Jotta voidaan alkaa tekemään dataprojektia, tulee ymmärtää mitä ollaan tekemässä ja mihin tarkoitukseen. Jos esimerkiksi ollaan luomassa koneoppimismallia koodiosastolle, niin ratkaisu eroaa todella paljon siitä, jos malli halutaan rakentaa asiakkaille käytettäväksi. Myös tulee ymmärtää liiketoimintaa sen verran, että projekti osataan suunnitella niin, että se tukee liiketoimintaa parhaalla mahdollisella tavalla. Kyseissä työssä ei tulla syventymään liiketoiminnan ymmärtämiseen, sillä projekti ei vastaa suoraan minkään liiketoiminnan ongelmaan.

#### Vaihe 2. 
Seuraavaksi prosessissa siirrytään datan hankintaan ja ymmärtämiseen. Datan hankinta on kyseisessä työssä tehty kevyesti valmiin API:n avulla, mutta monestikin dataprojekteissa se ei ole aina niin yksinkertaista. Data voi tulla esimerkiksi monista lähteistä ja sitä voidaa joutua muokkaamaan hyvinkin paljon ennen kuin se on hyödynnettävissä. Tässä vaiheessa tulee myös määritellä, miten dataa tullaan käsittelemään ja missä ympäristössä. Useasti modernit projektit rakennetaan pilviympäristöön. Ennen kuin data on vielä hyödynnettävissä, tulee se siivota ja mahdollisesti muokata hyödynnettävään muotoon. Tämä pitää sisällään esimerkiksi tyhjien arvojen poistamista sekä uusien sarakkeiden luomista, jos tarpeellista. Tulee myös ymmärtää, mitä data pitää sisällää, jotta siitä saadaan kaikki tarpeellinen irti. Tämä voi pitää sisällään esimerkiksi alustavia visualisointeja datan sisällöstä.

#### Vaihe 3.
Kun data on muokattu sellaiseksi, että sitä voidaan alkaa hyödyntämään, alkaa vaihe mallinnus. Liiketoiminnan tarve määrittelee, mitä tarpeeseen mallin tulee vastata, ja datan avulla sitä aletaan suorittamaan. Liiketoiminnan tarve tulee pitää mielessä jatkuvasti mallia rakentaessa. Malli voi olla esimerkiksi ohjatun oppimisen koneoppimisalgoritmi. Mallinnus voi pitää sisällään esimerkiksi ominaisuussuunnittelua (feature engineering), mallin koulutusta (Model training) ja mallin arviointia (Model Evaluation). Kyseisessä työssä on rakennettu 4 mallia datan pohjalta:
- Logistinen regressiomalli
- Lineaarinen regressiomalli
- NLP-malli
- Klusterointimalli

#### Vaihe 4.
Viimeisessä vaiheessa projekti viedään käyttöönottoon. Tätä vaihetta kyseinen työ ei pidä suoraan sisällää, mutta tämän datatiedeprojektin pohja voitaisiin periaatteessa viedä palautuksen jälkeen käyttöönottoon.

### Koneoppiminen - Ohjattu oppiminen

Kuten yllä mainittiin, ohjattu oppiminen perustuu ohjelmalle opetettuun koulutusaineistoon. Ohjattua oppimista käytetään yleisesti ennustavan mallinnuksen tekemisessä. Ennustava malli on malli, joka on rakennettu koneoppimisalgoritmin ja piirteiden tai attribuuttien avulla koulutusaineistosta, jonka avulla voidaan ennustaa tuleva arvo hyödyntämällä koulutuaineistoa.  Ohjatut oppimisalgoritmit pyrkivät mallintamaan suhteita ja riippuvuuksia kohde-ennusteen tuotoksen ja syötteiden piirteiden välillä. Käytetyimmät ohjatun oppimisen algoritmityypit ovat luokittelu ja regressio. 

Luokittelualgoritmi on ohjatun oppimisen menetelmä, jonka tavoitteena on luokitella uusia "arvoja" tiettyihin luokkiin. Luokittelualgoritmi saa opetusdatana joukon esimerkkejä, joista jokainen sisältää piirteitä ja luokan, johon se kuuluu. Algoritmi käyttää näitä esimerkkejä oppiakseen tunnistamaan piirteitä, jotka ovat ominaisia kullekin luokalle. Luokittelualgoritmi voi olla esimerkiksi:
- kasvojentunnistusalgoritmi
- sairauden diagnosointialoritmi
- sähköpostien luokittelualgoritmi

Regressioalgoritmissa tavoitteena on ennustaa numeerinen arvo muuttujan perusteella. Regressioalgoritmit hyödyntää opetusdataa oppiakseen ennustamaan piirteitä, jotka ovat yhteydessä riippuvan muuttujan arvoon. Niitä voidaan hyödyntää monissa eri asioissa kuten: 
- Ennustava analytiikka
- Markkinointitutkimus
- Sääennuste
- Taloudellinen ennustaminen

Seuraavaksi käydään ohjattua oppimista koodiesimerkkien avulla läpi. Ensimmäiseksi luokittelu (logistinen regressio) ja sitten regressio (lineaarinen regressio).

```python
# Hyödynnetään aiemmin muokattua test_df_01
test_df_01
```

```python
# Luokitellaan logistisen regression avulla data ja luodaan malli, joka ennustaa tuleeko match,
# kun tiedetään arvio omasta ulkonäöstä (1-10), arvio toisen ulkonäöstä (1-10) sekä molempien iät

# erotellaa muuttujat, poistetaan kommentit, sillä vielä tyhjiä
X_01 = test_df_01.drop('match', axis=1).drop('comments', axis=1)
y_01 = test_df_01['match']

# Jaetaan data koulutus- ja testaussetteihin
X_train_01, X_test_01, y_train_01, y_test_01 = train_test_split(X_01, y_01, test_size=0.3, random_state=42)

# Sovitetaan logistinen regressio harjoitusdataan
clf = LogisticRegression(random_state=42, max_iter=1000, solver='liblinear')
clf.fit(X_train_01, y_train_01)

# Tehdään ennuste testidatalle
y_pred_01 = clf.predict(X_test_01)

# arvioidaan mallin suoritusta

print('Accuracy:', accuracy_score(y_test_01, y_pred_01))
print('ROC AUC:', roc_auc_score(y_test_01, y_pred_01))
```

Nyt luotu malli arvioi noin 84% prosentin todennäköisyydellä oikein, tuleeko match. ROC AUC arvo on n. 0.52 mikä on lähellä satunnaisarvon (0.5) ennustuskykyä. Tämä tarkoittaa, että mallin ennustuskyky on hyvin heikko ja se ei pysty erottamaan positiivisia ja negatiivisia havaintoja kovin hyvin. Tämä kertoo siis sen, että on vaikeaa ennustaa annettujen lähtöarvojen perusteella, tuleeko match. Tehdään vielä vähän syvällisempää tutkailua tuloksista.

Kokeillaan hieman toisenlainen esimerkki käyttäen lineaarista regressiota. Hyödynnetään samaa dataa ja arvioidaan sen perusteella, mikä on arviotu matchien määrä.

```python
# Erotellaan muutujat ja huomioidaan testiin vain vastanneen osapuolen tulokset, jotta saadaan todellisempi kuva siitä, paljon tietty henkilö arvioi saavansa matcheja
X_02 = test_df_01.drop(['expected_num_matches','attractive_o','match','Asian/Pacific Islander/Asian-American_o',
       'European/Caucasian-American_o', 'Other_o', 'attractive_important',
       'Latino/Hispanic American_o', 'Black/African American_o','comments'], axis=1)
y_02 = test_df_01['expected_num_matches']

# Jaetaan koulutus- ja testausdataan
X_train_02, X_test_02, y_train_02, y_test_02 = train_test_split(X_02, y_02, test_size=0.2, random_state=42)

# Luodaan lineaarisen regression malli
lr = LinearRegression()

# Koulutuetaan malli koulutusdatalla
lr.fit(X_train_02, y_train_02)

# Tehdään ennustus
y_pred_02 = lr.predict(X_test_02)

# Arvioidaan suorituskykyä
mse = mean_squared_error(y_test_02, y_pred_02)
mae = mean_absolute_error(y_test_02, y_pred_02)
r2 = r2_score(y_test_02, y_pred_02)

print(f"Mean squared error: {mse:.2f}")
print(f"Mean absolute error: {mae:.2f}")
print(f"R-squared score: {r2:.2f}")
```

Mallin ennustamat arvot poikkeavat keskimäärin 0.31 (neliöjuuri MSE:stä) yksiköllä todellisesta arvosta neliöitynä sekä ennustamat arvot poikkeavat keskimäärin 0.19 yksiköllä todellisesta absoluuttisena arvona. #### Ohjaamaton oppiminen - Klusterointi

Klusterointi on menetelmä, jossa jaetaan datassa olevat havainnot eri ryhmiin (klustereihin) niiden samankaltaisuuksien perusteella. Klusteroinnilla pyritään löytämään mahdollisia rakenteita datassa, ja sitä käytetään monilla eri aloilla, kuten tieteessä, markkinoinnissa ja tekniikassa. Klusterointi on yleinen menetelmä ohjaamattomassa oppimisessa. Seuraavaksi muutama yleinen klusterointimetodi:
- Partitioning Methods: Menetelmä jakaa datan ennaltamäärättyyn määrän klustereita. Jokainen datapiste kuuluu vain yhteen klusteriin ja klusterien keskukset lasketaan iteratiivisesti. Kyseisessä työssä hyödynnetään tämän metodin menetelmää k-means.

- Hierarchical Methods: Menetelmä muodostaa hierarkian klustereista, aloittaen suuresta klusterista, joka sitten jaetaan pienempiin klustereihin, tai aloitetaan jokaisesta pisteestä omana klusterinaan ja sitten yhdistellään niitä hierarkisesti.

- Density-Based Methods: Menetelmä etsii alueita, joissa on korkea datapisteiden tiheys, joiden ympärillä on alueita, joissa on matala datapisteiden tiheys. 

- Grid-Based Methods: Menetelmä luo verkoston (grid), joka peittää datan ja muodostaa klustereita siten, että jokainen verkon solu on yksi klusteri.

- Model-Based Clustering Methods: Menetelmä käyttää tilastollista mallia klusterien löytämiseen. Näiden menetelmien tarkoituksena on löytää todennäköisyysjakauma, joka kuvaa dataa ja löytää sen piirteet. 

Seuraavaksi lähdetään muodostamaan työn datalle 5 klusteria kiinnostuksen kohteiden perusteella, hyödyntäen menetelmää k-means.

```python
# Skaalataan yksikkövarainssiksi
scaler = StandardScaler(with_mean=False)
X_scaled = scaler.fit_transform(test_df_02)

# Sovitetaan kmeans 5 klusterilla
kmeans = KMeans(n_clusters=5)
kmeans.fit(X_scaled)

# Lisätään clusteriluokka dataframeen
test_df_02['cluster'] = kmeans.labels_

# Tulostetaan klustereiden määrät
print(test_df_02['cluster'].value_counts())
```
